```python
import random

def hangman():
    word_list = ["python", "hangman", "programming", "developer", "algorithm"]
    word = random.choice(word_list).lower()
    guessed = ["_"] * len(word)
    attempts = 6
    guessed_letters = set()

    print("üéÆ Welcome to Hangman!")

    while attempts > 0 and "_" in guessed:
        print("\nWord:", " ".join(guessed))
        print(f"Attempts left: {attempts}")
        print("Guessed letters:", " ".join(sorted(guessed_letters)))
        guess = input("Enter a letter: ").lower()

        if not guess.isalpha() or len(guess) != 1:
            print("Please enter a single alphabet letter.")
            continue

        if guess in guessed_letters:
            print("You already guessed that letter.")
            continue

        guessed_letters.add(guess)

        if guess in word:
            for i, letter in enumerate(word):
                if letter == guess:
                    guessed[i] = guess
            print(f"‚úÖ Good guess: {guess}")
        else:
            attempts -= 1
            print(f"‚ùå Wrong guess: {guess}")

    print("\nGame Over!")
    if "_" not in guessed:
        print("üéâ Congratulations, you won!")
    else:
        print(f"üò¢ You lost. The word was '{word}'.")

# Run the game
hangman()
```
